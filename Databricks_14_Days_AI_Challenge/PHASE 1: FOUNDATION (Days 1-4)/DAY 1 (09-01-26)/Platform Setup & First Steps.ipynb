#Load Data Using PySpark (Basic Read)

```python
def load_ecommerce_data(spark, path: str):
    """
    Loads raw e-commerce CSV data using Spark.
    """
    return (
        spark.read
        .option("header", True)
        .option("inferSchema", True)
        .csv(path)
    )

df_events = load_ecommerce_data(spark, data_path)
```

## Return Valid Users

```python
from pyspark.sql.functions import col

def filter_valid_users(df):
    return df.filter(col("user_id").isNotNull())

df_filtered_valid = filter_valid_users(df_events)
df_filtered_valid.show(5)
```
